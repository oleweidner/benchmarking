# Process Profiling

```python
import psutil
from pylab import figure, show
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def measure_process(pid):
    """Measures the CPU time and memory consumption of the process 
       given by 'pid'.
    """
    netstat = psutil.net_io_counters(pernic=False)
    bytes_send_ref = netstat.bytes_sent
    bytes_recv_ref = netstat.bytes_recv
    
    cpu_samples = []
    mem_samples = []
    net_samples = []
    bytes_sent_samples = []
    bytes_recv_samples = []
    p = psutil.Process(pid)
    try: 
        while p.status() == 'running':
            cpu_samples.append(p.cpu_percent(interval=1.0))
            mem_samples.append(psutil.TOTAL_PHYMEM/1024/1024/100 * p.memory_percent())
            bytes_sent_samples.append(psutil.net_io_counters(pernic=False).bytes_sent - bytes_send_ref)
            bytes_recv_samples.append(psutil.net_io_counters(pernic=False).bytes_recv - bytes_recv_ref)
    except psutil.NoSuchProcess:
        pass
        
    return np.array([cpu_samples, mem_samples, bytes_sent_samples, bytes_recv_samples]) 
```

```python
def plot_data(data):
    """Plots the CPU usage and memory consumption data.
    """
    fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(16,8))
    ax[0][0].plot(data[0], 'r-')
    ax[0][0].set_ylabel('CPU Usage (%)')
    ax[0][0].set_xlabel('Runtime (seconds)')

    ax[0][1].plot(data[1], 'b-')
    ax[0][1].set_ylabel('Memory Usage (MB)')
    ax[0][1].set_xlabel('Runtime (seconds)')
    
    ax[1][0].plot(data[2], 'g-')
    ax[1][0].set_ylabel('Bytes sent')
    ax[1][0].set_xlabel('Runtime (seconds)')
    
    ax[1][1].plot(data[3], 'g-')
    ax[1][1].set_ylabel('Bytes received')
    ax[1][1].set_xlabel('Runtime (seconds)')
```
